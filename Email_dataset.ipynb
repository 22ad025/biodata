```python
data1['num_char']=data1['text'].apply(len)
data1.head(5)
```

```python
data1['num_words'] = data1['text'].apply(lambda x: len(nltk.word_tokenize(x)))
data1.head(5)
```

```python
data1['num_sentence'] = data1['text'].apply(lambda x: len(nltk.sent_tokenize(x)))
data1.head(5)
```

```python
data1['transformed_test']=data1['text']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(data1['transformed_test']).toarray()
x
```

```python
y=data1['details'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
```

```python
gnb=GaussianNB()
mnb=MultinomialNB()
```

```python
gnb.fit(x_train,y_train)
y_pred_gnb=gnb.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred_gnb))
print(confusion_matrix(y_test,y_pred_gnb))
```

```python
mnb.fit(x_train,y_train)
y_pred_mnb=mnb.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred_mnb))
print(confusion_matrix(y_test,y_pred_mnb))
```
